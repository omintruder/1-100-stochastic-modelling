```python
import math

#algorithmic block
def check_loop(sequence):
  for i in range(len(sequence)):
    for j in range(len(sequence)):
      if i != j and sequence[i] == sequence[j]:
        return 0
  return 1

def chaotic_function(a, sequence):
  while check_loop(sequence):
    if a < 50:
      a *= 2
    else:
      a = 2 * (100 - a)
    sequence.append(a)

#statistics block

#finds the mean
def find_average(slovar):
  sum = 0
  for key in slovar:
    sum += slovar[key]
  return sum / len(slovar)

#finds the longest cycle:
def find_max(slovar):
  ans = 0
  for key in slovar:
    if ans <= slovar[key]:
      ans = slovar[key]
  return ans

#upd: not used. might be used later
def check_deviations(avg, slovar):
  for key in slovar:
    if slovar[key] < avg:
      print(f'{key, slovar[key]}')

#upd: not used. might be used later
def deviations_list(max, slovar):
  deviations = {}
  for key in slovar:
    if slovar[key] <= max:
      deviations[key] = slovar[key]
  return deviations

#classifies distribution of cycles:
def reveal_deviations(max, deviations):
  dev_set = {}
  beta = []
  for i in range(1, max + 1):
    for key in deviations:
      if deviations[key] == i:
        dev_set[i] = beta.append(key)
    if len(beta) > 0:
      print(f'{i} steps to cycle for {len(beta)} values: {beta}')
      beta.clear()

#main body

stats = {}
for i in range(1, 100):
  list = []
  chaotic_function(i, list)
  stats[i] = len(list) - 1
  print(f'{i} ended at {list[len(list) - 1]}. cycle is {len(list) - 1} of elements:')
  print(list[0:(len(list) - 1)])

print(f'\n////////////////////////////////////// \n')
print(f'avg is {find_average(stats)} \nmax value is {find_max(stats)}.')
print(f'distribituion goes like: ')
reveal_deviations(find_max(stats), stats)
#check_deviations(find_average(stats), stats)
```
